```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
```

```python
import scipy.io.wavfile as wav
srate, audio = wav.read("amazing_grace_female.wav")
audio = audio.astype(np.float)/32767
plt.figure()
plt.plot(audio)
ipd.Audio(audio, rate = srate)
```

```python
mfcc = librosa.feature.mfcc(y=audio, sr=srate)
print("Audio size: %d" % audio.size)
print("MFCC dimesions: (%d,%0d)" % mfcc.shape)
print("MFCC size: %d" % mfcc.size)
print("Compression ratio: %2.2f%%" %  ((float(mfcc.size)/float(audio.size)) * 100))

naudio = librosa.feature.inverse.mfcc_to_audio(mfcc)
ipd.Audio(naudio,rate=srate)
```
